```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Tools

def create_scatter_plot(x, y, xlabel, ylabel, title):
    plt.scatter(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
    
def z_score_normalization(arr):
    arr_copy = arr.copy()
    arr_copy.sort()
    mu = arr_copy.mean()
    sigma = arr_copy.std()
    normalized_arr = (arr_copy - mu) / sigma
    return normalized_arr
```

```python
# This function takes a file containing gametocytemia data formatted accordingly
# and it returns tuple of arrays
def read_csv_exflag_vs_gams_data(filename):
    data = pd.read_csv(filename) # read csv into pandas DataFrame
    
    # Produce arrays for each of the columns containing gams data. Each col should be from a different source
    exflag_source1 = data['exflag_source1'].to_numpy()
    gams_source1 = data['gams_source1'].to_numpy()
    gams_source2 = data['gams_source2'].to_numpy()
    # gams_source3 = gams_data['gams_source3'].to_numpy()
    # gams_source4 = ...
    
    # Produce arrays of age and codenames for each of the measurements from each of the sources
    feed_date = data['feed_date'].to_numpy()
    codenames = data['codenames'].to_numpy()
    
    # Return these arrays in a tuple
    return feed_date, codenames, exflag_source1, gams_source1, gams_source2
```

```python
# Saves outputs as an age of gams array, a codename array, and tuple of measurement arrays of N length (N = number of sources) 

FILENAME = "../../data/p2_exflag_vs_gams.csv" # Set the path to your file here
feed_date, codenames, exflag_source1, gams_source1, gams_source2 = read_csv_exflag_vs_gams_data(filename=FILENAME) # Run function here
```

```python
# Simple plot of gametocytemia against exflagellation
create_scatter_plot(
    x=gams_source1, 
    y=exflag_source1, 
    xlabel="Gametocytemia (%)", 
    ylabel="Exflagellation (%)", 
    title="How does gametocytemia relate to exflagellation? \nSource 1 Data (Not Normalized)"
)

# Simple plot of gametocytemia against exflagellation
create_scatter_plot(
    x=gams_source2, 
    y=exflag_source1, 
    xlabel="Gametocytemia (%)", 
    ylabel="Exflagellation (%)", 
    title="How does gametocytemia relate to exflagellation? \nSource 2 Data (Not Normalized)"
)
```

```python
normalized_gams_source1 = z_score_normalization(gams_source1)
normalized_gams_source2 = z_score_normalization(gams_source2)
```

```python
create_scatter_plot(
    x=normalized_gams_source1, 
    y=exflag_source1, 
    xlabel="Gametocytemia (%)", 
    ylabel="Exflagellation (%)", 
    title="How does gametocytemia relate to exflagellation? \nSource 1 Data (Normalized)"
)

create_scatter_plot(
    x=normalized_gams_source2, 
    y=exflag_source1, 
    xlabel="Gametocytemia (%)", 
    ylabel="Exflagellation (%)", 
    title="How does gametocytemia relate to exflagellation? \nSource 2 Data (Normalized)"
)
```
